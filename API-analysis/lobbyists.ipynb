```python
import requests
import json
from config import congress_key
```

```python
lobby_url = "https://api.propublica.org/congress/v1/lobbying/latest.json"
lobby_r = requests.get(lobby_url, headers={"X-API-Key": congress_key})
lobby_json = lobby_r.json()
lobby_json
```

```python
lobbying_reps = lobby_json["results"][0]["lobbying_representations"]
lobbying_reps
```

```python
for rep in lobbying_reps:
    client = rep["lobbying_client"]["name"]
    registrant = rep["lobbying_registrant"]["name"]
    lobbyists = rep["lobbyists"]
    lobbyist_names = []
    for lobbyist in lobbyists:
        lobbyist_names.append(lobbyist["name"])
    date = rep["latest_filing"]["filing_date"]
    issues = rep["specific_issues"]
    report = rep["latest_filing"]["pdf_url"]
    print(f"Client: {client}")
    print(f"Registrant: {registrant}")
    print("Lobbyist name(s): ")
    print(*lobbyist_names, sep=", ")
    print("Issue(s): ")
    print(*issues, sep =", ")
    print(f"Latest filing date: {date}")
    print(f"Latest filing report (link): {report}")
    print("-"*8)
    
    
```
